## Wprowadzenie

[Quarto](https://quarto.org/) to darmowy system służący do tworzenia wysokiej jakości dokumentów, prezentacji, stron internetowych i raportów przy użyciu kodu oraz prostego tekstu.

W połączeniu z Jupyter Notebook Quarto działa jako warstwa rozszerzająca możliwości notebooków, umożliwiając przekształcenie analizy obliczeniowej w estetyczne, gotowe do publikacji materiały.

W praktyce Quarto pozwala pisać dokumenty w formacie opartym na Markdownie (pliki .qmd), w których można osadzać kod w Pythonie (lub innych językach). Kod jest wykonywany podobnie jak w Jupyterze, a jego wyniki — takie jak tabele, wykresy czy inne rezultaty — są automatycznie wstawiane do końcowego dokumentu.

Jedną z największych zalet Quarto jest oddzielenie treści od formy prezentacji. Ten sam dokument można wygenerować w różnych formatach, takich jak HTML, PDF, Word czy slajdy, bez konieczności zmiany treści. Dzięki temu narzędzie jest szczególnie przydatne w analizie danych, badaniach naukowych i edukacji.

Quarto integruje się z Jupyterem poprzez:
- uruchamianie kerneli Jupytera do wykonywania kodu,
- obsługę plików .ipynb,
- możliwość konwersji między notebookami a plikami .qmd,
- wspieranie reprodukowalnych analiz, gdzie wyniki są automatycznie odtwarzane.

Krótko mówiąc, Quarto rozszerza możliwości Jupytera — z interaktywnego środowiska do pisania kodu przekształca go w kompletne narzędzie do tworzenia profesjonalnych, powtarzalnych publikacji.

Nim zaczniemy, musimy zainstalować Quarto. Najłatwiej jest zrobić to poprzez odwiedzenie tego [linku](https://quarto.org/docs/get-started/).

Jeśli chcemy pracować w Visual Studio Code, możemy też po prostu zainstalować odpowiednią wtyczkę.

Dokument można wygenerować z poziomu konsoli (komenda *quarto render nazwa_pliku.qmd*), albo klikając odpowiedni przycisk w Visual Studio Code (prawy górny róg, *Preview*).

## Filozofia pracy z Quarto

Można powiedzieć, że Quarto jest czymś więcej niż narzędziem do tworzenia raportów - to sposób pracy stawiający na pierwszym miejscu:

1. powtarzalność, tzn. każdy wynik można odtworzyć,
2. śledzenie procesu, tzn. wiemy skąd pochodzą wyniki,
3. automatyzację raportowania, tzn. raport aktualizuje się wraz z danymi,
4. jednorodność źródła, tzn. kod, opis i wyniki w jednym miejscu.

W praktyce chodzi o to, żeby odejść od ręcznego kopiowania, "magicznych" wyników bez kodu, analiz których nie da się odtworzyć "po tygodniu".

## Quarto vs. Jupyter notebook

Zarówno Jupyter notebook jak i Quarto cechują się dużą interaktywnością. Niemniej jednak opcje raportowania w Jupyterze są bardzo ograniczone, tzn. możemy importować całą zawartość pliku .ipynb do .pdf albo .html, ale w zasadzie nie możemy ingerować w dokument w taki sposób, żeby cechowały go wysokie walory estetyczne. 

Quarto jest *de facto* odpowiedzią na potrzeby biznesowe, oferując narzędzie do tworzenia replikowalnych, skalowalnych raportów. Jupyter jest z kolei narzędziem dydaktycznym, eksploracyjnym, pozwalającym na pracę nad małymi projektami niekomercyjnymi.

## Podstawy pracy w *Markdown*

Dokumentacja dostępna jest [tutaj](https://quarto.org/docs/authoring/markdown-basics.html).

Markdown to lekki język znaczników, który pozwala formatować tekst w prosty i czytelny sposób. W Quarto jest on podstawowym sposobem tworzenia treści dokumentów.

### Nagłówki

Nagłówki w *Markdown* tworzymy w następujący sposób:

```markdown
# Nagłówek 1
## Nagłówek 2
### Nagłówek 3
```

Tu wyjątkowo nie pokażemy jak to wygląda już w .html, bo zepsułoby to spis treści i automatyczne numerowanie sekcji. Poza tym nagłówki są już nieodłącznym elementem tego dokumentu.

### Formatowanie tekstu

Standardowe formatowanie w *Markdown* wygląda tak w pliku .qmd:

```markdown
**pogrubienie**
*kursywa*
`kod inline`
```

a tak w wynikowym .html:

**pogrubienie**
*kursywa*
`kod inline`

Punktory w .qmd formułujemy tak:

```markdown
- Pierwszy punkt
- Drugi punkt
  - Podpunkt
```

a w .html wyglądają tak:

- Pierwszy punkt
- Drugi punkt
  - Podpunkt

Analogicznie numeracja, najpierw jak wygląda w .qmd:

```markdown
1. Pierwszy krok
2. Drugi krok
3. Trzeci krok
```

i jak wygląda już w .html:

1. Pierwszy krok
2. Drugi krok
3. Trzeci krok

### Wstawianie elementów w tekst

Z najpopularniejszych wariantów, umiejscawiamy w teście linki. Tak to wygląda po stronie .qmd:

```markdown
[Quarto](https://quarto.org)
```

a tak w .html:

[Quarto](https://quarto.org)

Możemy też wstawiać cytaty:

```markdown
> To jest przykład cytatu w Markdown.
```

> To jest przykład cytatu w Markdown.

Możemy też wstawiać "bloki" (*callouts*):

```markdown
::: {.callout-note}
To jest ważna informacja.
:::
```

::: {.callout-note}
To jest ważna informacja.
:::

```markdown
::: {.callout-warning}
Uwaga na typowy błąd!
:::
```

::: {.callout-warning}
Uwaga na typowy błąd!
:::

Często interesuje nas zapis matematyczny (LaTeX) na poziomie Quarto:

```markdown
$E = mc^2$
```

$E = mc^2$

```markdown
$$
\int_{0}^{1} x^2 dx
$$
```

$$
\int_{0}^{1} x^2 dx
$$

### Tabele

Składnia pozwalająca na tworzenie tabel od zera jest dość specyficzna. Mianowicie należy napisać to:

```markdown
| Default | Left | Right | Center |
|---------|:-----|------:|:------:|
| 12      | 12   |    12 |   12   |
| 123     | 123  |   123 |  123   |
| 1       | 1    |     1 |   1    |

: Demonstration of pipe table syntax
```

żeby uzyskać to:

|---------|:-----|------:|:------:|
| 12      | 12   |    12 |   12   |
| 123     | 123  |   123 |  123   |
| 1       | 1    |     1 |   1    |

W praktyce raczej konstruujemy tabele na poziomie wybranego języka programowania i pracujemy nad estetycznym umieszczeniem ich w *Markdown*. Do tego zagadnienia też wrócimy za moment, serdeczna prośba o cierpliwość.

### Diagramy

Konstrukcja diagramów sprowadza się do wykorzystania specyficznej, choć mimo wszystko dość intuicyjnej składni:

```markdown
flowchart LR
  A[Hard edge] --> B(Round edge)
  B --> C{Decision}
  C --> D[Result one]
  C --> E[Result two]
```

umieszczanej wewnątrz symboli *```* ze specyfikacją narzędzia służącego do tworzenia takich struktur, często stosowany jest oparty o Javascript *{mermaid}*. Wynik prezentuje się następująco:

```{mermaid}
flowchart LR
  A[Hard edge] --> B(Round edge)
  B --> C{Decision}
  C --> D[Result one]
  C --> E[Result two]
```

Alternatywą jest specyfikacja *{dot}* oznaczająca narzędzie grpahviz. Poniżej przykład składni tegoż i wynik:

```markdown
graph G {
  layout=neato
  run -- intr;
  intr -- runbl;
  runbl -- run;
  run -- kernel;
  kernel -- zombie;
  kernel -- sleep;
  kernel -- runmem;
  sleep -- swap;
  swap -- runswap;
  runswap -- new;
  runswap -- runmem;
  new -- runmem;
  sleep -- runmem;
}
```
